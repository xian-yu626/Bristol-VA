```python
import pandas as pd

data_path = '2021WORK.csv'
data = pd.read_csv(data_path)

data.head(), data.shape
```

```python
import umap.umap_ as umap

umap_model = umap.UMAP(n_components=2, random_state=42)

numerical_data = data.iloc[:, 2:]

embedded_data = umap_model.fit_transform(numerical_data)

umap_df = pd.DataFrame(embedded_data, columns=['UMAP_1', 'UMAP_2'])
umap_df['Area'] = data['Area']  # Adding Area for reference

# Save
umap_output_path = 'UMAP_2021WORK.csv'
umap_df.to_csv(umap_output_path, index=False)

umap_df.head(), umap_output_path
```

```python
import matplotlib.pyplot as plt

plt.scatter(umap_df['UMAP_1'], umap_df['UMAP_2'])
plt.title('UMAP')
plt.xlabel('UMAP_1')
plt.ylabel('UMAP_2')
plt.show()
```
